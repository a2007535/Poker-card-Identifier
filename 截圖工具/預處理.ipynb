```python
import cv2
import numpy as np
import os

# 假設你的數據集存放在一個名為 'dataset' 的資料夾中，並有子資料夾分別包含不同撲克牌的圖片。
# 例如，'dataset/Ace', 'dataset/2', 'dataset/3', 等等。

data_path = 'screenshot/'
categories = [i+j for i in 'A23456789TJQK' for j in "shdc"]

# 定義一個函數來進行圖片的增強
def augment_image(img):
    # 上下翻轉
    flipped_ud = cv2.flip(img, 0)
    # 左右翻轉
    flipped_lr = cv2.flip(img, 1)
    # 旋轉90度
    rotated_90 = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
    # 旋轉180度
    rotated_180 = cv2.rotate(img, cv2.ROTATE_180)
    # 旋轉270度
    rotated_270 = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)

    # 返回增強後的圖片集
    return [flipped_ud, flipped_lr, rotated_90, rotated_180, rotated_270]

# 遍歷每個類別的資料夾
for category in categories:
    path = os.path.join(data_path, category)
    
    # 遍歷每張圖片
    for img_name in os.listdir(path):
        img_path = os.path.join(path, img_name)
        img = cv2.imread(img_path, cv2.IMREAD_COLOR)
        # 執行圖片增強
        augmented_images = augment_image(img)
        
        # 將增強後的圖片保存到相同的資料夾
        for idx, augmented_img in enumerate(augmented_images):
            new_img_name = f"{img_name.split('.')[0]}_aug_{idx}.jpg"
            new_img_path = os.path.join(path, new_img_name)
            cv2.imwrite(new_img_path, augmented_img)
```

```python
import cv2
import numpy as np
import os

# 假設你的數據集存放在一個名為 'dataset' 的資料夾中，並有子資料夾分別包含不同撲克牌的圖片。
# 例如，'dataset/Ace', 'dataset/2', 'dataset/3', 等等。

data_path = 'screenshot/'
categories = [i+j for i in 'A23456789TJQK' for j in "shdc"]

# 定義一個函數來進行圖片的增強
def augment_image(img, target_size=(70, 70)):
    # 計算填充的大小
    rows, cols, chennals = img.shape
    top = int((target_size[0] - rows) / 2)
    bottom = target_size[0] - rows - top
    left = int((target_size[1] - cols) / 2)
    right = target_size[1] - cols - left
    # 填充圖片
    img = cv2.copyMakeBorder(img, top, bottom, left, right, cv2.BORDER_REPLICATE)

    # 旋轉15~360度
    rotated = []
    rows, cols, chennals = img.shape
    for angle in range(0,360,90):
        rotation_matrix = cv2.getRotationMatrix2D((cols / 2, rows / 2), angle, 1)
        rotated.append(cv2.warpAffine(img, rotation_matrix, (cols, rows)))
        
    # # 旋轉90度
    # rotated_90 = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
    # # 旋轉180度
    # rotated_180 = cv2.rotate(img, cv2.ROTATE_180)
    # # 旋轉270度
    # rotated_270 = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)

    # 返回增強後的圖片集
    return rotated

# 遍歷每個類別的資料夾
for category in categories:
    path = os.path.join(data_path, category)
    
    # 遍歷每張圖片
    for img_name in os.listdir(path):
        img_path = os.path.join(path, img_name)
        img = cv2.imread(img_path, cv2.IMREAD_COLOR)
        
        # 執行圖片增強
        augmented_images = augment_image(img)
        
        # 將增強後的圖片保存到相同的資料夾
        new_path = os.path.join('preProcess/', category)
        if not os.path.exists(new_path):
            os.makedirs(new_path)

        for idx, augmented_img in enumerate(augmented_images):
            new_img_name = f"{img_name.split('.')[0]}_aug_{idx}.jpg"
            new_img_path = os.path.join(new_path, new_img_name)
            cv2.imwrite(new_img_path, augmented_img)
```
