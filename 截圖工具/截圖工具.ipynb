```python
import tkinter as tk
from tkinter import filedialog
from PIL import ImageGrab, ImageTk
import os
from datetime import datetime
import cv2


def getTheSelectPosition():
    ImageGrab.grab(all_screens=True).save("tmp.jpg")
    img = cv2.imread('tmp.jpg')

    dot1 = []
    dot2 = []
    img2 = img
    def show_xy(event, x, y, flags, param):
        nonlocal dot1, dot2, img, img2  # 使用nonlocal讓函數內部能夠修改外部的變數
        if flags == 1:
            if event == 1:
                dot1 = [x, y]
            if event == 0:
                img2 = img.copy()
                dot2 = [x, y]
                cv2.rectangle(img2, (dot1[0], dot1[1]), (dot2[0], dot2[1]), (0, 0, 255), 1)
                cv2.imshow('ScreenPositionSelect', img2)
            if event == 4:
                cv2.destroyAllWindows()  # 關閉視窗
                return

    cv2.namedWindow("ScreenPositionSelect", cv2.WINDOW_NORMAL)
    cv2.setWindowProperty("ScreenPositionSelect", cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)
    cv2.imshow('ScreenPositionSelect', img)
    cv2.setMouseCallback('ScreenPositionSelect', show_xy)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    return dot1, dot2


class ScreenshotTool:
    def __init__(self, master):
        self.master = master
        self.master.title("自動截圖工具")

        # 三個按鈕
        self.select_area_button = tk.Button(self.master, text="選擇截圖範圍", command=self.start_select_area)
        self.select_area_button.pack(pady=10)

        self.save_button = tk.Button(self.master, text="儲存截圖", command=self.save_screenshot)
        self.save_button.pack(pady=10)

        # 輸入框
        self.category_var = tk.StringVar()
        self.category_entry = tk.Entry(self.master, textvariable=self.category_var, width=20)
        self.category_entry.pack(pady=10)
        self.category_label = tk.Label(self.master, text="輸入分類名稱")
        self.category_label.pack()
        self.suit_label = tk.Label(self.master, text="s 黑桃 h 紅心 d 方塊 c 梅花")
        self.suit_label.pack()

        # 顯示截圖的區域
        self.image_label = tk.Label(self.master)
        self.image_label.pack()

        # 兩個輸入框
        self.start_x_var = tk.StringVar()
        self.start_x_entry = tk.Entry(self.master, textvariable=self.start_x_var, width=10)
        self.start_x_entry.pack(side=tk.LEFT, padx=5)
        self.start_x_label = tk.Label(self.master, text="左上角X座標")
        self.start_x_label.pack(side=tk.LEFT)

        self.start_y_var = tk.StringVar()
        self.start_y_entry = tk.Entry(self.master, textvariable=self.start_y_var, width=10)
        self.start_y_entry.pack(side=tk.LEFT, padx=5)
        self.start_y_label = tk.Label(self.master, text="左上角Y座標")
        self.start_y_label.pack(side=tk.LEFT)

        self.end_x_var = tk.StringVar()
        self.end_x_entry = tk.Entry(self.master, textvariable=self.end_x_var, width=10)
        self.end_x_entry.pack(side=tk.LEFT, padx=5)
        self.end_x_label = tk.Label(self.master, text="右下角X座標")
        self.end_x_label.pack(side=tk.LEFT)

        self.end_y_var = tk.StringVar()
        self.end_y_entry = tk.Entry(self.master, textvariable=self.end_y_var, width=10)
        self.end_y_entry.pack(side=tk.LEFT, padx=5)
        self.end_y_label = tk.Label(self.master, text="右下角Y座標")
        self.end_y_label.pack(side=tk.LEFT)

        # 拖曳選擇範圍的變數
        self.dragging = False
        self.start_x = 0
        self.start_y = 0
        self.end_x = 0
        self.end_y = 0

    def start_select_area(self):
        # 開始拖曳選擇範圍
        self.master.update()
        dot1, dot2 = getTheSelectPosition()
        print(dot1, dot2)
        self.start_x = dot1[0]
        self.start_y = dot1[1]
        self.end_x = dot2[0]
        self.end_y = dot2[1]
        self.update_coordinates()

        # 顯示截圖
        screenshot = ImageGrab.grab(bbox=(self.start_x, self.start_y, self.end_x, self.end_y))
        self.show_screenshot(screenshot)

    def update_coordinates(self):
        # 更新輸入框中的座標
        self.start_x_var.set(str(self.start_x))
        self.start_y_var.set(str(self.start_y))
        self.end_x_var.set(str(self.end_x))
        self.end_y_var.set(str(self.end_y))

    def save_screenshot(self):
        # 這個函數用於儲存截圖

        # 獲取分類名稱
        category = self.category_var.get()

        # 如果資料夾不存在，則創建
        if not os.path.exists("screenshot"):
            os.makedirs("screenshot")
        if not os.path.exists(f"screenshot/{category}"):
            os.makedirs(f"screenshot/{category}")

        # 儲存截圖到 screenshot 資料夾
        time = str(datetime.now().time()).replace(":","-")
        screenshot_path = f"screenshot/{category}/{time}.png"
        screenshot = ImageGrab.grab(bbox=(self.start_x, self.start_y, self.end_x, self.end_y))
        screenshot.save(screenshot_path)

    def show_screenshot(self, screenshot):
        # 將截圖轉換為Tkinter PhotoImage格式
        screenshot_tk = ImageTk.PhotoImage(screenshot)

        # 在Label中顯示截圖
        self.image_label.config(image=screenshot_tk)
        self.image_label.image = screenshot_tk

if __name__ == "__main__":
    root = tk.Tk()
    app = ScreenshotTool(root)
    root.mainloop()
```
